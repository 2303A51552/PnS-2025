```python
import numpy as np
import matplotlib.pyplot as plt
# Data
daily_calories = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise_duration = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])
```

```python
# Scatter Plots
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(daily_calories, bmi)
plt.title("Calories vs. BMI")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 2)
plt.scatter(exercise_duration, bmi)
plt.title("Exercise vs. BMI")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 3)
plt.scatter(daily_calories, exercise_duration)
plt.title("Calories vs. Exercise")
plt.show()
```

```python
# Pearson Correlation
print("Calories vs. BMI:", np.corrcoef(daily_calories, bmi)[0, 1])
print("Exercise vs. BMI:", np.corrcoef(exercise_duration, bmi)[0, 1])
print("Calories vs. Exercise:", np.corrcoef(daily_calories, exercise_duration)[0, 1])
```
